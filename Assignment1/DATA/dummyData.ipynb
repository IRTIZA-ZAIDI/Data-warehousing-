```python
import csv
import random
import datetime
from faker import Faker
import pandas as pd
```

```python
fake = Faker()
```

```python
def random_date(start, end):
    return fake.date_between(start_date=start, end_date=end)
```

```python
# User table
user_data = []

# Create 10 admins
for _ in range(10):
    user_data.append(
        [
            _,  
            fake.user_name(),
            fake.email(),
            fake.password(),
            "admin",
        ]
    )

for _ in range(10, 1000):
    role = random.choice(["seller", "customer"])
    user_data.append(
        [
            _,
            fake.user_name(),
            fake.email(),
            fake.password(),
            role,
        ]
    )
```

```python
# Save User data to CSV
with open("users.csv", "w", newline="") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(
        ["user_id", "username", "email", "password", "role"]
    )  # Column headers
    writer.writerows(user_data)
```

```python
# Assuming df_products has the necessary product details
df_products = pd.read_csv("product_data.csv")

# Load category data
df_category = pd.read_csv("category_data.csv")

# Create a list to store product-category associations
product_category_list = []

# Initialize index counter
index_counter = 0

# Iterate through each product row
for idx, row in df_products.iterrows():
    # Generate a random number between 1 and 3 to represent the number of categories
    num_categories = random.randint(1, 3)

    # Choose unique random categories for the product
    chosen_categories = random.sample(list(df_category["category_id"]), num_categories)

    # Create entries for each product-category association
    for category_id in chosen_categories:
        product_category_list.append(
            {
                "product_id": row["product_id"],
                "category_id": category_id,
                "product_category_id": index_counter,
            }
        )
        index_counter += 1

# Create the new DataFrame 'df_product_category'
df_product_category = pd.DataFrame(product_category_list)
df_product_category.set_index(
    "product_category_id", inplace=True
)  # Set index to 'product_category_id'

# Save the new DataFrame to a CSV file named 'product_category.csv'
df_product_category.to_csv("product_category.csv")

# Display the new DataFrame
print("Generated Product_Category Data:")
print(df_product_category)
```
